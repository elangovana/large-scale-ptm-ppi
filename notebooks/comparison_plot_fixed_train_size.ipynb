```python
from scipy import stats

def plot_stdline(x, y_mean, y_std, ax, label, color, no_label=False):
    if no_label:
        label=None
    ax.plot(x, y_mean, color=color, label=label, linewidth = 4.0, marker='o')
    ax.fill_between(x, y_mean - y_std, y_mean + y_std, color=color, alpha=0.4)
    ax.grid(True)
    ax.set_title(label)


def plot_tvalue(df, r, y_column,  ax, x_col, alternative, color="black"):
    t_test = stats.ttest_ind(df.query(f"{x_col} == {r}")[y_column], 
                df.query(f"{x_col} == {0.0}")[y_column], 
                alternative=alternative)
    if str(t_test.pvalue) == "nan" :
        ax.text(r, 82, "***"
             , color= color, rotation="vertical")
    elif t_test.pvalue <=0.001:
        ax.text(r, 82, "***"
             , color= color, rotation="vertical")
    elif t_test.pvalue <=0.05:
        ax.text(r, 82, "**"
             , color= color, rotation="vertical")

    elif t_test.pvalue <0.1:
        ax.text(r, 82, "*"
             , color= color, rotation="vertical")

def plot_comparison_linestd(df, ax, x_col, **kwargs):
    
    with warnings.catch_warnings():
        warnings.filterwarnings("ignore", message=".*")

        my_colors= [ "green",  "orange"]
        
        linetype = "-"
        legend_suffix = ""


        column_renames={"Precision": "Precision" , 
                        "TrainSize" : "Train size" ,
                        "Recall" : "Recall" , 
                        
                        "F1": "F1" , 
                       }
        ax[0].set_ylim(0,105)
        ax[0].set_xlim(0,0.82)


        ldf = df.groupby(x_col)["Precision"].describe()
        plot_stdline(ldf.index, ldf["mean"], ldf["std"], ax[0],  label='Precision',  **kwargs)
       

        ax[1].set_ylim(0,105)
        ax[1].set_xlim(0,0.82)
        ldf = df.groupby(x_col)["Recall"].describe()
        plot_stdline(ldf.index, ldf["mean"], ldf["std"], ax[1],  label='Recall',  **kwargs)


        ax[2].set_ylim(0,105)
        ax[2].set_xlim(0,0.82)
        ldf = df.groupby(x_col)["F1"].describe()
        plot_stdline(ldf.index, ldf["mean"], ldf["std"], ax[2],  label='F1',  **kwargs)

        
```

```python

color = "lightblue"
```

```python
def plot_comparisons(df_chem, df_selfsup, x_col, xlabel, alternative, significancecolor):
    fig, ax = plt.subplots(2,3, figsize=(27,10))
    plot_comparison_linestd(df_chem, ax[0],x_col=x_col , color=color)
    plot_comparison_linestd(df_selfsup, ax[1], x_col=x_col, color=color, no_label=True)
    # plot_comparison_linestd(df_affables, ax, color="lightgreen",)


    ax[1][1].set_xlabel(xlabel)

    axi=ax[0]
    axi[0].set_ylabel("ChemProtC3")
    max_l = int(df_chem[x_col].max()*10)+1


    for i in range(1,max_l):
        plot_tvalue(df_chem, i/10, "Precision",axi[0], x_col, color=significancecolor, alternative=alternative)

    for i in range(1,max_l):
        plot_tvalue(df_chem, i/10, "Recall", axi[1], x_col, color=significancecolor, alternative=alternative)


    for i in range(1,max_l):
        plot_tvalue(df_chem, i/10, "F1", axi[2], x_col, color=significancecolor, alternative=alternative)




    axi=ax[1]
    axi[0].set_ylabel("SelfSupKD")
    max_l = int(df_selfsup[x_col].max()*10) + 1
    for i in range(1,max_l):
        plot_tvalue(df_selfsup, i/10, "Precision", axi[0], x_col, color=significancecolor, alternative=alternative)

    for i in range(1,max_l):
        plot_tvalue(df_selfsup, i/10, "Recall", axi[1], x_col, color=significancecolor, alternative=alternative)


    for i in range(1,max_l):
        plot_tvalue(df_selfsup, i/10, "F1", axi[2],x_col,  color=significancecolor, alternative=alternative)





```

```python
plot_comparisons(df_affablepos_chemprot, df_affablepos_selfsup, 
                 "TrainAffablePositiveRate", 
                 xlabel="$r_\mathrm{hf}^{p}$",
                 alternative ="less", 
                 significancecolor = "coral" )
plt.show()
```

```python
plot_comparisons(df_affableneg_chemprot, df_affableneg_selfsup, 
                 "TrainAffableNegativeRate", 
                 xlabel="$r_\mathrm{hf}^{n}$",
                 alternative ="less", 
                 significancecolor = "coral" )
plt.show()
```
